### **Outliers Values**

Tahap ini dilakukan agar kita dapat melihat nilai ekstrem (outliers) pada kolom yang bertipe data kan numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

Tidak semua kolom perlu ditangani, tetapi kita tetap harus melihat dan memahami tujuan dari setiap kolom yang ada di dataset.

Terlihat pada kolom **rating** memiliki persentase sebanyak 1%
dan terlihat pada kolom **total_reviews** memiliki persentase sebanyak 13%

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

Aktivitas: Membandingkan nilai rating (rating) dengan total review (total_reviews)

Tujuan: Mengidentifikasi dan membandingkan total nilai rating yang diberikan oleh pengunjung

Visualisasi: Bar Chart (Grafik Batang).

```python
perbandingan_rating = df.groupby('rating')['total_reviews'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,10))
sns.barplot(x=perbandingan_rating.index, y=perbandingan_rating.values, palette='viridis', hue=perbandingan_rating.index, legend=False)
plt.title('Perbandingan Nilai Tempat Wisata')
plt.xlabel('Rating')
plt.ylabel('Total Reviews')
plt.xticks(rotation=45)
plt.show()
```

## **Composition/Komposisi**

Aktivitas: Melihat proporsi atau persentase total reviews berdasarkan provinsi (province).

Tujuan: Mengetahui provinsi yang paling mendominasi total reviews secara keseluruhan.

Visualisasi: Pie Chart (Grafik Lingkaran).

```python
rating_by_province = df.groupby('province')['total_reviews'].sum().sort_values(ascending=False)
rating_by_province_top3 = rating_by_province.head(3)

plt.figure(figsize=(10, 8))
rating_by_province_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 3 Provinsi Teratas Berdasarkan Total Reviews')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

## **Distribution/Distribusi**

Aktivitas: Menganalisis penyebaran lokasi wisata pada kolom ( province).

Tujuan: Melihat apakah mayoritas transaksi bernilai kecil atau besar, serta mendeteksi adanya kecondongan (skewness) dalam data penjualan.

Visualisasi: Histogram.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['province'], bins=20, kde=True)
plt.title('Penyebaran Lokasi (Province)')
plt.xlabel('Province')
plt.ylabel('Frekuensi')
plt.show()
```

## **Relationship/Hubungan**

Aktivitas: Menganalisis korelasi data antara **(total_reviews)** dan **(rating)**

Tujuan: Mengetahui apakah total review tempat wisata berbanding lurus dengan rating pada tempat wisata.
Visualisasi: Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['total_reviews', 'rating']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara rating dan total_review')
plt.show()
```
