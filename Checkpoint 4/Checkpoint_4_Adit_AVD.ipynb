### **Outliers Values**

Tahap ini dilakukan agar kita dapat melihat nilai ekstrem (outliers) pada kolom yang bertipe data kan numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

Tidak semua kolom perlu ditangani, tetapi kita tetap harus melihat dan memahami tujuan dari setiap kolom yang ada di dataset.

Terlihat pada kolom **rating** memiliki persentase sebanyak 1%
dan terlihat pada kolom **total_reviews** memiliki persentase sebanyak 13%

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Outliers Values**

Dalam dataset, terkadang ada nilai yang jauh berbeda dari nilai lainnya dalam dataset, nilai itu adalah Outliers. Nilai Outlier bisa jauh lebih rendah atau lebih tinggi. Jika data outliers ada <= 10% dari banyaknya
jumlah data maka dapat dihapus, tetapi jika >= 10% bisa saja diabaikan

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

1. **rating**

Kolom rating menunjukkan persentase outliers sebanyak 1%. Ini dapat disebabkan karena penilaian suatu tempat tidak selalu konsisten dan merupakan hal yang wajar. Jadi ini dapat diabaikan karena kita membutuhkannya.

2. **total_reviews**

Kolom **total_reviews** memiliki persentase outliers sebanyak 13%. Bagian ini juga wajar karena dalam dataset jumlah total review yang diperoleh suatu tempat sangat bervariasi tergantung dari popularitasnya. Jadi outlier ini dapat diabaikan.

# **Construct Data**

Construct Data adalah bagian dimana kita membuat data baru dalam dataset dengan menggunakan data yang sudah ada. Tujuannya adalah agar analisis yang dilakukan dapat memberikan informasi lebih yang dibutuhkan.

  **1. Membuat Kolom Baru "Popularity"**
  
Kolom Popularity berfungsi sebagai indikator jika suatu tempat wisata merupakan tempat yang populer atau tidak.

```python
def popularity_category(totalreviews):
    if totalreviews > 0:
        return 'Populer'
    elif totalreviews < 0:
        return 'Kurang Populer'
    else:
        return 'Tidak Ada Informasi'

df['popularity'] = df['total_reviews'].apply(popularity_category)
```
